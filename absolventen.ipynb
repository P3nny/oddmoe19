```python
data = json.load(open('abschluesse_gesamt_ohne.15.json'))
```

```python
df = pd.DataFrame(data["data"]["allRegions"]["regions"])
```

```python
df.head()
```

```python
df["BILABS"][0]
```

```python
def convert(df):
    for _, row in df.iterrows():
        row['Gesamt Weiblich'] = None
        row['Gesamt Männlich'] = None
        row['Ohne Abschluß Weiblich'] = None
        row['Ohne Abschluß Männlich'] = None
        for line in row['BILABS']:
            line['id'] = row['id']
            line['name'] = row['name']
            if line['GES'] == 'GESW':
                if row['Gesamt Weiblich'] == None:
                    row['Gesamt Weiblich'] = line['value']
                    row['Ohne Abschluß Weiblich'] = line['value']
                elif row['Gesamt Weiblich'] < line['value']:
                    row['Gesamt Weiblich'] = line['value']
                else:
                    row['Ohne Abschluß Weiblich'] = line['value']
            elif line['GES'] == 'GESM':
                if row['Gesamt Männlich'] == None:
                    row['Gesamt Männlich'] = line['value']
                    row['Ohne Abschluß Männlich'] = line['value']
                elif row['Gesamt Männlich'] < line['value']:
                    row['Gesamt Männlich'] = line['value']
                else:
                    row['Ohne Abschluß Männlich'] = line['value']
        yield row
```

```python
df = pd.DataFrame(convert(df))
df.head()
```

```python
df.columns
```

```python
df1 = df[['id', 'name', 'Gesamt Weiblich', 'Gesamt Männlich',
       'Ohne Abschluß Weiblich', 'Ohne Abschluß Männlich']]
```

```python
df1.head()
```

```python
df1['Gesamt'] = df1['Gesamt Weiblich'] + df1['Gesamt Männlich']
```

```python
df1['ohne Abschluss'] = df1['Ohne Abschluß Weiblich'] + df1['Ohne Abschluß Männlich']
```

```python
df
```

```python
df1 = df1[['id', 'name', 'Gesamt', 'ohne Abschluss']]
```

```python
df1['Anteil ohne Abschluss'] = (df1['ohne Abschluss'] / df1['Gesamt'])*100
```

```python
df1 = df1[['name', 'Anteil ohne Abschluss']]
```

```python
df1 = df1.dropna()
```
